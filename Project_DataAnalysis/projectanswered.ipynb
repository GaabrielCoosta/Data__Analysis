```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
total = df.Total.value_counts()
print(f"Total complaints registered: {total}")
```

```python
# Second Question
max = df["Tempo Resposta"].max()
print(f"O tempo máximo: {max}")

min = df["Tempo Resposta"].min()
print(f"Tempo mínimo: {min}")

med = df["Tempo Resposta"].mean()
print(f"Tempo médio: {med}")
```

```python
# Third question
nmax = df["Nota do Consumidor"].max()
print(f"Nota máxima: {nmax}")

nmin = df["Nota do Consumidor"].min()
print(f"Nota mínima: {nmin}")

nmed = df["Nota do Consumidor"].mean()
print(f"Nota média: {nmed}")
```

```python
# Sixth Question
import matplotlib.pyplot as plt
df = pd.read_csv("1-dadosgovbr---2014.csv",sep = ';', encoding="latin-1")
n= df[["UF","Total"]].groupby("UF").count()
print(n)

```

```python
# Seventh Question
registro = df['Respondida']
nregistro = 0
sregistro = 0


for x in registro:
    if x == 'N':
        nregistro = nregistro + 1
    if x == 'S':
        sregistro = sregistro + 1
tregistro = nregistro + sregistro        
pnregistro = (nregistro*100) / tregistro
print(f"Porcentagem de reclamações não respondidas: {pnregistro:.2f}%")
```

```python
# Fifth Question
# a) Frequencia de reclamações por sexo
```

```python
import matplotlib.pyplot as plt
df = pd.read_csv("1-dadosgovbr---2014.csv",sep = ';', encoding="latin-1")
plt.style.use('seaborn-paper')
plt.rcParams['figure.figsize'] = (11,7)
plt.title('Frequência de Reclamações por Sexo',fontsize=16)
graf = df[["Sexo","Total"]].groupby("Sexo").count()
plt.bar(graf.index, height=graf.Total, color=['purple','grey'], width = 0.5)
plt.xlabel('Sexo',fontsize=12)
plt.ylabel('Quantidade',fontsize=12)
plt.show()
```
